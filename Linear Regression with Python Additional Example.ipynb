```python
%%time
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from linreg import LinearModel
```

```python
#Explicitly specified the xlrd engine
df = pd.read_excel('chirps.xls', engine='xlrd')
df.head()
```

```python
x = df.X.values
y = df.Y.values
x = np.reshape(x, (x.shape[0], 1))
y = np.reshape(y, (y.shape[0], 1))
```

```python
model = LinearModel(1)

losses = model.train(
    x, y,
    200, 0.0001
)
```

```python
plt.plot(losses);
```

```python
y_preds = model.forward_pass(x)
```

```python
plt.plot(y_preds, y, 'r.')
plt.plot(y_preds, y_preds, 'b-', label='Best fit')
plt.xlabel('Predicted Y')
plt.ylabel('Actual Y')
plt.legend()
plt.show()
```
